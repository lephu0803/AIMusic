```python
import pandas as pd
```

```python
df = pd.read_csv("./five_chord_prob.csv")
```

```python
df
```

```python
prob = dict()
prob[''] = []
def find_prob(x):
    global prob
    roman_cp = x['roman_cp'].split(',')
    proba = x['probability'].split(',')
    for i in range(len(roman_cp) - 1):
        if i == 0:
            prob[''].append((roman_cp[i],proba[i]))
            prob[''] = list(set(prob['']))
        else:
            if roman_cp[i] not in prob.keys():
                prob[roman_cp[i]] = []
            prob[roman_cp[i]].append((roman_cp[i + 1],proba[i + 1]))
            prob[roman_cp[i]] = list(set(prob[roman_cp[i]]))
df.apply(lambda x : find_prob(x),axis=1)
```

```python
prob
```

```python
con_prob = pd.DataFrame(columns=['cond','target','prob'])
```

```python
con_prob
```

```python
for k,v in prob.items():
    for t in v:
        con_prob = con_prob.append({
            "cond" : k,
            "target" : t[0],
            "prob" : t[1]
        },ignore_index=True)
```

```python
con_prob.to_csv("./chord_cond_markov.csv",index=False)
```
